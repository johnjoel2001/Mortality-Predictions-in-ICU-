```python
columns=['ALP', 'ALT', 'AST', 'Age', 'Albumin', 'BUN', 'Bilirubin',
       'Cholesterol', 'Creatinine', 'DiasABP', 'FiO2', 'GCS', 'Gender',
       'Glucose', 'HCO3', 'HCT', 'HR', 'Height', 'ICUType', 'K', 'Lactate',
       'MAP', 'MechVent', 'Mg', 'NIDiasABP', 'NIMAP', 'NISysABP', 'Na',
       'PaCO2', 'PaO2', 'Platelets', 'RecordID', 'RespRate', 'SaO2', 'SysABP',
       'Temp', 'TroponinI', 'TroponinT', 'Urine', 'WBC', 'Weight', 'pH','In-hospital_death']
```

```python
def outlier(df3,column):
    upper_limit=df3[column].mean()+3*df3[column].std()
    lower_limit=df3[column].mean()-3*df3[column].std()
    
    ls=df3.index[(df3[column]<lower_limit)|(df3[column]>upper_limit)]
    return ls
```

```python
index_ls=[]
for column in columns:
    index_ls.extend(outlier(df3,column))
```

```python
index_ls
```

```python
def remove(df3,ls):
    df3=df3.drop(ls)
    return df3
```

```python
df_new=remove(df3,index_ls)
df_new.shape
```

###  Building the model

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
from tensorflow.keras.layers import Dropout,BatchNormalization
from tensorflow.keras.optimizers import Adam
```

```python
# model=keras.Sequential([
#     keras.layers.Dense(26,input_shape=(X_train.shape[1],),activation='relu'), #Input layer
#     keras.layers.Dense(15,activation='relu'),  #Hidden layer
#     keras.layers.Dense(2,activation='sigmoid') #Output layer
# ])

model = Sequential()

model.add(Dense(64, input_dim=X_train.shape[1] , activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(196, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(200, activation='relu'))
model.add(Dropout(0.5))

model.add(BatchNormalization())

model.add(Dense(256, activation='relu'))
model.add(Dense(2, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, Y_train , epochs=200 , batch_size = 128 , validation_data=(X_test, Y_test))
```

```python
print(history.history.keys())
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
